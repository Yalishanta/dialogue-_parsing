```python
import numpy as np
import json
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import random
import matplotlib.ticker as ticker
from scipy.stats import norm

df =  pd.read_csv("test_data.csv")
df
```

### удаляяем все строки с репликами "client", т.к. нас интересует толькот то что сказал менеджер

```python
drop_df = df.set_index('role')
drop_df.drop(['client'],axis = 0, inplace = True)
drop_df = drop_df.reset_index() 
drop_df
```

### Извлекать реплики с приветствием – где менеджер поздоровался. 
В столбце "greetings" показано поздоровался ли manager или нет "greetings_list" - это список выражений которые мог использовать manager в данной ситуации

```python
test_list =['здравствуйте', 'добрый день', 'приветствую вас', 'доброе утро', 'добрый вечер', 'привет', 'мое почтение', 'здрасте', 'здравствуй']
ft_greetings = []
for i in range(len(drop_df)):
    test_string = drop_df.iloc[i]['text'].lower()
    res = [ele for ele in test_list if(ele in test_string)]
    T_or_F = str(bool(res))
    ft_greetings.append(T_or_F)
drop_df["greetings"] = ft_greetings 
drop_df
```
